## <u>Arithmetic Precedence</u>

dffd.jpeg

**📖 Do you Know 📖**

In Python, the order of operations follows the PEMDAS rule, which is an acronym to help remember the sequence of operations:

1. **P - Parentheses**:
    - Operations enclosed in parentheses are performed first. Nested parentheses are evaluated from the innermost level to the outermost.
2. **E - Exponents**:
    - After parentheses, exponentiation is performed next. This includes powers and square roots.
3. **M - Multiplication** and **D - Division**:
    - These operations are performed next and are of equal precedence. They are executed from left to right. Division includes true division (**`/`**), floor division (**`//`**) and Modulo (**%**) . (You'll learn about the differences soon enough).
4. **A - Addition** and **S - Subtraction**:
    - Finally, addition and subtraction are performed, also from left to right. They are of equal precedence.

Python adheres to this order of operations, which is consistent with the standard mathematical conventions. Understanding PEMDAS is crucial for writing correct expressions in Python, as it dictates how complex expressions will be evaluated and in what sequence the operations will occur.

```python
# Multipl arithmetic operator together in place

2 + 3 * 4 # Multiplication has to be done before addition | 14 not 20
```

```python
(2+3) * 4 # Bracket comes the first in level of importance.
```

```python
3 + 4 - 5 + 6
# Addition & Substraction : Both are on the same level | Whichever comes first will be executed from the left side
# First : 3+4 = 7 |--> 7 - 5 + 6
# Second : 7-5 = 2 | 2+ 6
# Third : 2 + 6 = 8
```

```python
3 + 4 - 5 * 2 + 6
# First : 5 * 2 = 10 | 3 + 4 - 10 + 6 :: Multiplcation has the higher importance than addition or substraction
# Second : 3 + 4 =7 |7 -10 + 6
# Third : 7 - 10 =-3 | -3 + 6
# Finally -3 + 6 = 3
```

```python
3 + 4 * 2 - 5 % 2 + 6
# First : 4 * 2 = 8 | 3 + 8 - 5 % 2 + 6
# Second : 5 % 2 = 1 | 3 + 8 - 1 + 6
# Third : 3 + 8 = 11 | 11 -1 + 6
# Fourth : 11- 1 = 10 | 10 + 6
# Finally 10 +6 = 16
```
