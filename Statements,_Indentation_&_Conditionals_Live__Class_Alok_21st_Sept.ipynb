```python
# if Statement:

# The if statement checks a condition.
# If the condition is True, the code block under the if statement is executed.
# Example:
age = int(input('Enter the age:')) # 16
if age >= 18 : # 16 >= 18 : False
  print('You are an adult')



# Explanation:
# Here, if the age is greater than or equal to 18, the message "You are an adult." is printed.
```

```python
# else Statement:

# The else statement is used to execute a block of code when the condition in the if statement is False.
# Example:
age = int(input('Enter the age:'))

if age >= 18 :
  print('You are an adult')
else:
  print('You are not an adult')


# Explanation:
# If age is not greater than or equal to 18, the message "You are not an adult." is printed.
```

```python
# elif Statement:

# The elif (short for "else if") statement allows you to check multiple conditions.
# It only runs if the previous if or elif condition was False.
# Example:

score = int(input('Enter your score'))

if score >=90:
  print('Grade A')
elif score >= 80 and score < 90:
  print('Grade B')
elif score >=70 and score <80:
  print('Grade C')
else:
  print('Grade D')

# Explanation:
# >=90 : Grade A
# >=80 < 90 : GradeB
# >=70 < 80 : Grade C
# If neither condition is true, "Grade D" is printed.
```

```python
score = int(input('Enter your score'))
if score >=90: # False
  print('Grade A')
if score >= 80 and score < 90: # True
  print('Grade B')
if score >=70 and score <80: # False
  print('Grade C')
else:
  print('Grade D')
```

```python
# Nested if-else Statements:

# Nested if-else statements are if-else structures inside another if-else.
# This allows you to check for more complex conditions.
# Example:
num = int(input('Enter the number:')) # 4
if num >= 0:# True
  if num ==0: # False
    print('Zero')
  else:
    print('Positive')
else:
  print('Negative')

# Explanation: If num is greater than or equal to 0, another if checks if num is 0.
# If true, "Zero" is printed. Otherwise, "Positive Number" is printed.
# If num is less than 0, "Negative Number" is printed.
```

```python
num = int(input('Enter the number:'))

if num ==0:
  print('Zero')
elif num >0 :
  print('Positive')
else:
  print('Negative')
```
